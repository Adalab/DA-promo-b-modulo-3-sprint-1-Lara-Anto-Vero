> Nos hemos dado cuenta que algunas columnas son de tipo object, y nos interesa cambiarlas a tipo numérico para poder operar con ellas. 

- bore
- stroke
- horsepower 
- peak-rpm

```python
df.select_dtypes(include='object')
```

```python
#Nos hemos dado cuenta de que tenemos valores '?' por lo que no podemos hacer un casteo. 
# Decidimos reemplazar esos valores por nulos para posteriormente cambiar el valor por la mediana
```

```python
col_interr=list(df.columns)

for i in col_interr:
    df[i].replace('?', np.nan, inplace=True)
```

```python
df[['bore','stroke','horsepower','peak-rpm']]=df[['bore','stroke','horsepower','peak-rpm']].astype(float, errors='ignore')
```

```python
df.dtypes
```

```python
df[['bore','stroke','horsepower','peak-rpm']].describe().T  #Decidimos reemplazar los nulos de las columnas que acabamos de castear por la mediana, ya que es el valor más fiable por el que podemos reemplazar, 
# habiendo observado que la media y la mediana no coinciden en valores. 
```

```python
df.isnull().sum()
```

```python
col_nan=['bore','stroke','horsepower','peak-rpm','normalized-losses']
```

```python
for i in col_nan:
    mediana=df[i].median()
    df[i]=df[i].fillna(mediana)
```

```python
df.isnull().sum()  #Ya on tenemos nulos en las variables categóricas con nulos
```
